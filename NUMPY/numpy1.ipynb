CREATING DIFFERENT LISTS 

```python
# example

list1 = [1,3,5,7,9]
list2 = [2,4,6,8,10] 

# we try multiply these elements together

list1 * list2
```

solution to the error

```python
new_list = []

for i in range(0, len(list1)):
    x = list1[i] * list2[i]
    new_list.append(x)
    
    print(new_list)
```
